```python
import yfinance as yf

import numpy as np
import pandas as pd

#define the ticker symbol
tickerSymbol = 'NVDA'

#get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

#get the historical prices for this ticker
df = tickerData.history(period='1d', start='2020-1-1', end='2022-10-6')

#see your data
df
```

```python
#define the ticker symbol
tickerSymbol = 'AMZN'

#get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

#get the historical prices for this ticker
df1 = tickerData.history(period='1d', start='2020-1-1', end='2022-10-6')

#see your data
df1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%pylab inline

pylab.rcParams['figure.figsize'] = (15, 9)   # Change the size of plots

df["Close"].plot(label='Amazon', color = '#000000')
df["Close"].plot(label='NVDA', color = '#FF00BF')

# titlu 
plt.title('Close price plot for Amazon & NVDA')

# label to the x-axis
plt.xlabel('Date')

# legend to the curve
plt.legend()
```

```python
close_AMZN = df["Close"]
close_NVDA = df1["Close"]

simple_return_AMZN = close_AMZN.pct_change()
simple_return_NVDA = close_NVDA.pct_change()

log_return_AMZN = np.log(1+simple_return_AMZN)
log_return_NVDA = np.log(1+simple_return_NVDA)
```

```python
import matplotlib.pyplot as plt   # Import matplotlib
# This line is necessary for the plot to appear in a Jupyter notebook
%matplotlib inline
# Control the default size of figures in this Jupyter notebook
%pylab inline

pylab.rcParams['figure.figsize'] = (15, 9)   # Change the size of plots

#apple["Adj. Close"].plot(grid = False) # Plot the adjusted closing price of AAPL

simple_return_AMZN.plot()# Plot the lsimple return
simple_return_NVDA.plot()# Plot the lsimple return


log_return_AMZN.plot() # Plot the logreturn
log_return_NVDA.plot() # Plot the logreturn
```

```python
plt.plot(simple_return_AMZN,log_return_AMZN, 'o',c='blue')
plt.ylabel( 'Log-return' )
plt.xlabel( 'Simple return' )
#plt.legend(loc='upper center',bbox_to_anchor=(0.5, -0.15), fancybox=True, shadow=True)
plt.show()
plt.plot(simple_return_NVDA,log_return_NVDA, 'o',c='green')
plt.ylabel( 'Log-return' )
plt.xlabel( 'Simple return' )
#plt.legend(loc='upper center',bbox_to_anchor=(0.5, -0.15), fancybox=True, shadow=True)
plt.show()
```
